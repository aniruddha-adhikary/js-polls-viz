## Initialize the Variables

```python
YEAR = 2008
```

## Load Dataframe

```python
import pandas as pd
import os

state_geo = os.path.join('data', 'districts_amended.json')
data_path = os.path.join('data', 'bangladeshelectiondata', 'Bangladesh{}ElectionResults.csv'.format(YEAR))

state_data = pd.read_csv(data_path)
```

## Cleanse Data

Remove voterTurnout over 100%. (?)

```python
state_data = state_data[state_data['voterTurnout']<=100]
state_data = state_data[['district', 'voterTurnout']]
```

## Aligning Data

We are awesome with spelling often.

```python
replacements = {
    'Chadpur': 'Chandpur',
    'Brahmminbaria': 'Brahamanbaria',
    'Hobiganj': 'Habiganj',
    'Jhalakathi': 'Jhalokati',
    'Laxmipur': 'Lakshmipur',
    'Moulvibazar': 'Maulvibazar',
    'Netrokona': 'Netrakona',
    'Narshingdi': 'Narsingdi',
    'Panchaghar': 'Panchagarh',
    'Sirajgonj': 'Sirajganj',
    "Mymensingh-Netrokona": 'Netrakona',
    "MymensinghNetrokona": 'Netrakona',
    'Barisal-Pirojpur': 'Pirojpur',
    'BarisalPirojpur': 'Pirojpur',
    'Borguna': 'Barguna',
    'Brahmminbaria': 'Brahamanbaria',
    'Jhalokathi': 'Jhalokati',
    'Khagrachari': 'Khagracchari',
    'Panchaghar': 'Panchagarh',
    'Shatkhira': 'Satkhira',
    'Sunamgonj': 'Sunamganj',
    'Khagrachari': 'Khagracchari'
}

for key, value in replacements.items():
    state_data['district'] = state_data['district'].replace(key, value)

state_data = state_data.groupby(['district']).mean().reset_index()
```

## Write to CSV

```python
state_data.to_csv('{}.csv'.format(YEAR))
```

## Draw Map

Center to Bangladesh, load a blank white photo as the tiles. And the rest is documented in folium.

```python
import folium

m = folium.Map(location=[23.6, 90.3],
               zoom_start=7,
               tiles='https://www.husseygaybell.com/wp-content/uploads/2018/05/blank-white-image.png',
               attr='Geo Data: BBS. Visualization: Aniruddha Adhikary')

folium.Choropleth(
    geo_data=state_geo,
    name='choropleth',
    data=state_data,
    columns=['district', 'voterTurnout'],
    key_on='feature.id',
    fill_color='Greens',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Voter Turnout (%) in {} JS Election'.format(YEAR),
    bins=5
).add_to(m)
```

## Saving the Map

Yup.

```python
m.save('{}.html'.format(YEAR))
```
